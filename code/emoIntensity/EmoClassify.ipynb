```python
def calculate_emotion_score(tokens, eemotion_words):
    total_score = 0
    count = 0
    for word in tokens:
        if word in eemotion_words:
            total_score += eemotion_words[word]
            count += 1
    return total_score / max(1,count)
```

```python
def classify_emotion(tokens, eemotion_words, word_embeddings):
    emotion_score = calculate_emotion_score(tokens, eemotion_words)
    
    token_vectors = [word_embeddings[word] for word in tokens if word in word_embeddings]

    mean_vector = np.mean(token_vectors, axis=0)
    if emotion_score >= 0.8:
        return "Very Emotional"
    elif emotion_score >= 0.6:
        return "Emotional"
    elif emotion_score >= 0.4:
        return "Neutral"
    elif emotion_score >= 0.2:
        return "Slightly Emotional"
    else:
        return "Not Emotional"
```

```python
import pandas as pd
df=pd.read_csv('train_emo.csv')
```

```python
import string

def punctuation_removal(text):
    all_list = [char for char in text if char not in string.punctuation]
    clean_str = ''.join(all_list)
    return clean_str
```

```python
df['text'] = df['text'].apply(punctuation_removal)
```

```python
sentences = df['text'].values
```

```python
word2vec_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=0)
```

```python
text = "Amidst the turmoil and anger in the world, finding moments of love, happiness, and contentment can bring a sense of relief and hope, allowing us to experience the full spectrum of human emotions"
tokens = punctuation_removal(text).split(" ")
print(len(tokens))
```

```python
word_embeddings = {word: word2vec_model.wv[word] for word in tokens if word in word2vec_model.wv}
```

```python
sentences_with_emotions = [
    "love and affection make life beautiful and meaningful.",
    "happiness and joy filled their hearts as they celebrated their success.",
    "The peaceful sunset by the beach brought a sense of serenity.",
    "In moments of contentment, one finds solace and purpose.",
    "surprise and amazement washed over her as he proposed unexpectedly.",
    "anger and frustration can sometimes cloud our judgment.",
    "sadness and grief overwhelmed her when she heard the news.",
    "fear and anxiety gripped him as he faced the unknown challenges.",
    "hatred and animosity can perpetuate cycles of conflict and hatred.",
    "despair and hopelessness took hold in the darkest of times.",
    "nostalgia and longing for the past can be bittersweet.",
    "A sense of relief and gratitude washed over her when she found her lost ring.",
    "A feeling of amusement and laughter filled the room as they watched a comedy show.",
    "He couldn't hide his enthusiasm and excitement for the upcoming adventure.",
    "amidst the chaos, a moment of calm and tranquility can be precious.",
    "pride and a sense of accomplishment swelled within him as he received the award.",
]
```

```python
for i in sentences_with_emotions:
    tokens = i.split(" ")
    emotion_level = classify_emotion(tokens, emotional_strength_dictionary, word_embeddings)
    print(f"The emotion level is: {emotion_level}")
```
